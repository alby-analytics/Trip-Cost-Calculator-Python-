```python
def hotel_cost(nights):
    cost_per_night = 140
    total_cost = nights * cost_per_night
    return total_cost

def plane_ride_cost(city):
    city = city.lower()  
    
    if city == "charlotte":
        return 183
    elif city == "tampa":
        return 220
    elif city == "pittsburgh":
        return 222
    elif city == "los angeles":
        return 475
    else:
        return 0
    
def rental_car_cost(days):
    base_cost = days * 40   
    
    # Discount logic
    if days >= 7:
        base_cost -= 50
    elif days >= 3:
        base_cost -= 20
    
    return base_cost

def trip_cost(city, days, spending_money):
    total = hotel_cost(days) \
            + plane_ride_cost(city) \
            + rental_car_cost(days) \
            + spending_money
    return total

city =str(input("Enter the city name").lower().strip())
days =int(input("Enter the number of night stay/days in hotel"))
spending_money =int(input("Enter spending money: "))

total_trip_cost = trip_cost(city, days, spending_money)

print("Total number of night stay in hotel =",days)
print("Total hotel cost =",hotel_cost(days))
print("City name:",city)
print("Round-trip flight cost =",plane_ride_cost(city))
print("Number of rental days =", days)
print("Final rental cost =",rental_car_cost(days))
print("Spending Money: $", spending_money)
print("Total Trip Cost: $", total_trip_cost)
```
